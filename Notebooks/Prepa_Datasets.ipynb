```python
df_arrondissement['Code_Commune'] = df_arrondissement['Code_Commune'].astype(str)
```

```python
df_arrondissement['Code_Commune'] = df_arrondissement['Code_Commune'].apply(lambda x: x.zfill(3))
```

```python
df_arrondissement['Code_Commune'] = df_arrondissement['Code_Commune'].apply(lambda x: '76' + x)
```

```python
df_arrondissement = df_arrondissement.rename(columns={'Code_Commune': 'com_insee'})
```

```python
categorie = ['Num_Acc', 'vma','catr','circ','nbv','surf']
```

```python
df_lieux_select = df_lieux[categorie]
```

```python
df_lieux_select.head()
```

```python
print(f'Nombre de lignes dupliquées : {df_lieux_select.duplicated().sum()}')
```

```python
df_lieux_select = df_lieux_select.drop_duplicates()
```

```python
print(f'Nombre de lignes dupliquées : {df_lieux_select.duplicated().sum()}')
```

```python
df_lieux_select.head(10)
```

```python
df = df_lieux_select.groupby('Num_Acc').last()
```

```python
# Lier lieux à caracteristiques
df_accidents_France = pd.read_csv(pathData + 'Accidents_France_2005_to_2023.csv',low_memory=False)
```

```python
df_accidents_France[(df_accidents_France.dep == 76) & (df_accidents_France.com_insee == '76540')]
```

```python
df.head(10)
```

```python
data_Accidents_lieux = df_accidents_France.merge(df, how='inner', on='Num_Acc')
```

```python
data_Accidents_lieux_76 = data_Accidents_lieux[data_Accidents_lieux['dep'] == 76]
```

```python
data_Accidents_lieux_76 = data_Accidents_lieux_76.merge(df_arrondissement,how='inner', on='com_insee')
```

```python
data_Accidents_lieux_76_2023 = data_Accidents_lieux_76[data_Accidents_lieux_76['an'] == 2023]
```

```python
data_Accidents_lieux_76_2023.shape
```

```python
# Sauvegarder CSV
data_Accidents_lieux_76.to_csv(pathData +  'data_Accidents_lieux_76.csv', index=False)
```
