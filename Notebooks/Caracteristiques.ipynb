```python
#Import librairies
import pandas as pd
import numpy as np
```

```python
# Chargement dataset Global
df_caracteristiques =pd.read_csv(pathData + 'caracteristiques_2005_to_2023.csv', sep=',', low_memory=False)

```

```python
print(f"Données caracteristiques: Nombre de lignes {df_caracteristiques.shape[0]} nombre de colonnes {df_caracteristiques.shape[1]}")
```

```python
# Majuscule pour le département concerne la corse 2a en 2A et 2b en 2B
df_caracteristiques['dep'] = df_caracteristiques['dep'].str.upper()
#correspondance_departement_region['departmentCode'] = correspondance_departement_region['departmentCode'].str.upper()
```

```python
# Renommer la colonne 'departmentCode' en 'dep'
correspondance_departement_region = correspondance_departement_region.rename(columns={'departmentCode': 'dep'})
```

```python
df_caracteristiques["dep"].unique()
```

```python
# Conversion Corse en 20
df_caracteristiques['dep'].mask( df_caracteristiques['dep'] == '2A', '20' , inplace=True )
df_caracteristiques['dep'].mask( df_caracteristiques['dep'] == '2B', '20' , inplace=True )
```

```python
df_caracteristiques.shape
```

```python
df_caracteristiques.dropna(subset=['dep'],inplace=True)
```

```python
df_caracteristiques = df_caracteristiques.astype({'dep':'int'})
```

```python
#supression des accidents Outre-mer
accidents_outre_mer = list(df_caracteristiques.loc[(df_caracteristiques.dep > 950),'Num_Acc'])
df_caracteristiques = df_caracteristiques[df_caracteristiques["Num_Acc"].isin(accidents_outre_mer) == False]
```

```python
df_caracteristiques.shape
```

```python
df_caracteristiques["dep"] = [_/10 if _ > 99 else _ for _ in df_caracteristiques["dep"].values]
```

```python
df_caracteristiques = df_caracteristiques.astype({'dep':'str'})
df_caracteristiques["dep"] = ["00"+_ if len(_) < 2 else "0"+_ if len(_) < 3 else _ for _ in df_caracteristiques["dep"].values]
```

```python
df_caracteristiques.info()
```

```python
df_caracteristiques.head(10)
```

```python
df_caracteristiques = df_caracteristiques.astype({'com':'str'})
df_caracteristiques["com_insee_temp"] = ["00"+_ if len(_) < 2 else "0"+_ if len(_) < 3 else _ for _ in df_caracteristiques["com"].values]
df_caracteristiques["com_insee_temp"] = df_caracteristiques["dep"] + df_caracteristiques["com_insee_temp"]
```

```python
df_caracteristiques["com_insee"] = df_caracteristiques["com_insee_temp"].str[-5:]
```

```python
df_caracteristiques.head(10)
```

```python
# Transformation du dataset correspondance région départements
# Remplissage des valeurs vides
correspondance_departement_region.regionName = correspondance_departement_region.regionName.fillna("")
correspondance_departement_region.regionCode = correspondance_departement_region.regionCode.fillna(0)
# Conversion en numérique
correspondance_departement_region['regionCode'] = correspondance_departement_region['regionCode'].astype(int)
```

```python
correspondance_departement_region.head(10)
```

```python
correspondance_departement_region["departmentCode"].unique()
```

```python
correspondance_departement_region["departmentCode"] = ["0"+_ if len(_) < 3 else _ for _ in correspondance_departement_region["departmentCode"].values]
```

```python
# Conversion Corse en 20
correspondance_departement_region['departmentCode'].mask( correspondance_departement_region['departmentCode'] == '02a', '020' , inplace=True )
correspondance_departement_region['departmentCode'].mask( correspondance_departement_region['departmentCode'] == '02b', '020' , inplace=True )
```
